```python
import os
import tarfile 
from six.moves import urllib

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

#FUNCION PARA RECUPERAR LOS DATOS DE UNA URL
def fetch_housing_data(housing_url = HOUSING_URL, housing_path = HOUSING_PATH):
    if not os.path.isdir(housing_path):                   #Verifica si existe el directorio especificado
        os.makedirs(housing_path)                         #Crea el directorio como mkdir
    tgz_path = os.path.join(housing_path, 'housing.tgz')  #Concatena los paths de manera legible para el OS
    urllib.request.urlretrieve(housing_url, tgz_path)     #Ejecuta el request de housing_url y lo guarda en tgz_path
    housing_tgz = tarfile.open(tgz_path)                  #Abre, extrae y cierra el archivo comprimido
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
```

# Crear el set de test

Vamos a crear el set de prueba, tomando aleatoriamente el 20% de los datos, es importante que no veamos este set de test para evitar el ***data snooping bias***, sesgo de espionaje de datos, y que provoque un mal desempeo del sistema en produccion.

```python
import numpy as np

def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data)) #Genera len(data) indices con un orden aleatorio, si se le pasara un array permuta aleatoriamente los elementos del array
    test_set_size = int(len(data) * test_ratio)   #Devuelve un numero entero con el valor de la longitud de los datos * test_ratio 0.2 para 20%, es decir el numero de valores que corresponden al 20%
    test_indices = shuffled_indices[:test_set_size]  #Guardo en una lista el primer 20% de lo indices generados
    train_indices = shuffled_indices[test_set_size:] #Guardo en una lista el resto despues del 20% de los indices
    return data.iloc[train_indices], data.iloc[test_indices] #Devuelvo 2 listas, una con los indices de entrenamiento y otra con los indices de test
#iloc es posicion basada en indices enteros, devuelve en un DF las filas a las que apuntas con el arreglo que le pasas
```

- La funcion generada arriba puede causar problemas si es llamada otra vez, porque al generar aleatoriamente los indices, para una segunda ejecucion el set de entramiento y de test, serian diferentes y con el paso del tiempo nuestro algorimo de ML, veria todo el dataset.
- Una solucion es guardar el primer set de test en la primer ejecucion y despues cargarlo en las subsecuentes.
- Otra opcion es establecer la semilla del random ` np.random.seed(42)` antes de llamar `random.permutation()`, de esta forma va a generar siempre los mismos indices

Pero a traves del tiempo cuando se actualice el dataset ambas opciones van a romperse.

- Una solucion comun es usar cada identificador de las instancias para decidir si deberia ir o no en el set de test (asumiendo que las instancias tienen un unico e inmutable identificador).

Por ejemplo: se puede generar un hash para cada una de las instancias, y poner la instancia en el set de test si el hash es menor o igual al 20% del maximo valor del hash.

Esto aseguraria que el set de test va a permanecer consistente a traves de multiples ejecuciones, incluso si refrescas el dataset.

El nuevo set de test contendra 20% de las nuevas instancias, pero no contendra ninguna instancia que estuvo previamente en el set de entrenamiento.

```python
#Esta seria una posible implementacion

from zlib import crc32

#Lo que hace esta funcion es dado un identificador y un porcentaje expresado en decimal
#verifica si el identificador se encuentra por debajo del 20% de todos los valores posibles
# que es 2^32
#es decir genera un checksum del identificador y verifica si este es menor al 20% de 2^32
#Devuele true o false
def test_set_check(identifier, test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32    #crc32(data[, value])
#Computa un CRC (Cyclic Redundancy Check) cheksum de data
#El resultado es un unsigned 322 bit integer
#Si value existe, es usado como el valor inicial del checksum, de otra forma se usa 0.
#Cambios en la nueva version siempre regresa un valor sin signo, para generar siempre el mismo valor
#a traves de todas las versiones de ptyhon y plataformar usar crc32(data) & 0xffffffff

def split_train_test_by_id(data, test_ratio, id_column):
    ids = data[id_column]   #Recupero el valor de la columna especificada en id_column
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio)) 
    return data.loc[~in_test_set], data.loc[in_test_set]

#En la linea 20 lo que se hace es que a cada valor de la columna guardad en ids se le aplica una una funcion
#esta funcion es generada de forma anonima con lambda donde el argumento es id_ y la expresion es la funcion
#definida para verificar si el checksum es menor o mayor. Entonces en in_test_set te genera una columna de true y false

#Finalmente la funcion retorna 2 data frame, uno que contenia los indices indicados con true y otro con los indices
#que indicaban false
```

Como desafortunadamente en este ejemplo los datos no tienen una columna como identificador unico, la manera mas simple es usar la columna index como el ID

```python
#Entonces generamos un dataframe que contenga el index sin afectar el original
housing_with_id = housing.reset_index()
train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "index")
```

Al usar la indice de la fila necesitamos asegurarnos que los nuevos datos que se agreguen deben aadirse al final del dataset, y que ninguna fila sea eliminada. Si no tenemos que buscar otro atributo que nos permita hacer un identificador unico.

Hasta este punto han sido metodos de muestreo completamente aleatorios, esto funciona bien si el dataset es lo suficientemente grande en relacion al numero de atributos, pero si no, se corre el riesgo de introducir un sesgo de muestreo.

#### Muestreo Estratificado

Cuando, por ejemplo,de una poblacion de 40% hombres y 60% mujeres, al tomar una muestra queremos asegurarnos que es representativa de la poblacion original, entonces queremos mantener la proporcion, a esto se le conoce como **muestreo estratificado**(***stratified sampling***). La poblacion es dividida en grupos llamados **estratos**(***strata***).

El numero correcto de instancias son muestreadas de cada **estrato** para garantizar que el set de test es representativo de toda la poblacion.

Nota: Si se hiciera con un muestreo completamente aleatorio existiria la posibilidad de muestrear un set de test sesgado, y por lo tanto los resultados tambien lo estarian

### Sobre el ejemplo

Supongamos que se nos menciona que el atributo `median_icnome` es muy importante para predecir los precios de los alojamientos, entonces nosotros queremos asegurarnos nuestro set de test es represantativo de las varias categorias de ingresos en todo el dataset.

Ya que `median_income` es un atributo numerico, necesitamos crear un atributo categorico de los ingresos.

Observando el histograma podemos darnos cuenta que la mayoria de los valores de ingresos se encuentran entre 1.5 y 6, pero algunos de los ingresos medios van mas alla del 6.

Es importante tener un suficiente numero de instancias en el dataset para cada estrato, si no de otra forma la importancia del estrato sera sesgada.

Esto significa que no debemos tener demasiados estratos, y cada estrato debe ser lo suficientemente grande.

```python
#Se usa una funcion para crear el atributo de categoria de ingreso, con 5 categorias del 1 al 5,
#donde la categoria 1 va de 0 a 1.5 y asi sucesivamente.

housing["income_cat"] = pd.cut(housing["median_income"],
                              bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                              labels = [1, 2, 3, 4, 5])

#pd.cut() nos funciona cuando queremos segementar y ordenar valores de datos en bins, tambien es util,
#cuando queremos ir de una variables continua a una variable categoria.
housing["income_cat"].hist()
```

Una vez hecho esto podemos a pasar a realizar el **muestreo estratificado** basado en la categoria de ingresos.
Para esto podemos hacer uso de la clase `StratifiedSuffleSplit` de Scikit-Learn

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
#n_splits indica el numero de iteraciones que se re-barajea y divide
#test_size el tamao de nuestro set de test
#random_state Semilla para el random
#Retorna los indices de entramiento/test para dividir los datos en sets de entrenamiento/test


#El metodo .split(self, X, y, groups=None) Genera indices para dividir los datos en los sets de entramiento/test
#X es un array, son los datos en entrenamiento
#Y es el target variable para problemas de aprendizaje supervidaso. 
#La estratificacion se hace basada en las etiquetas de Y
for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
    
#Aqui contamos los valores por cada categoria y sacamos su proporcion,
#es decir el porcentaje de datos de cada categoria    
strat_test_set["income_cat"].value_counts() / len(strat_test_set)
```

```python
#Proporcion del dataset original
housing["income_cat"].value_counts() / len(housing)
```

```python
def income_cat_proportions(data):
    return data["income_cat"].value_counts() / len(data)

train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)

compare_props = pd.DataFrame({
    "Overall": income_cat_proportions(housing),
    "Stratified": income_cat_proportions(strat_test_set),
    "Random": income_cat_proportions(test_set),
}).sort_index()
compare_props["Rand. %error"] = 100 * compare_props["Random"] / compare_props["Overall"] - 100
compare_props["Strat. %error"] = 100 * compare_props["Stratified"] / compare_props["Overall"] - 100
```

```python
compare_props
#Con esta funcion podemos ver una comparacion entre como estan categorizados todos los datos
#Comparando Todos los datos / los estratificados / los generados aleatoriamente /
#Error de hacerlo aleatoriamente / Error de manera estritificada
```

Con la tabla anterior podemos darnos cuenta que los datos estratificados tienen menor error a si lo hacemos de manera puramente aleatoria donde esta sesgado.

Una vez hecho la separacion de los set de test y entrenamiento, dbebemos eliminar la columa `income_cat`, que solo se uso de manera auxiliar, de esta forma regresamos los datos a su forma original.

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True)
```
